```python
#Q10.Average selling price by car model
print("\n 10. Avg Selling Price by Car:\n", df.groupby('Car_Name')['Selling_Price'].mean().sort_values(ascending = False))
```

-> Explanation:

â€¢ df.groupby('Car_Name') groups the dataset by each unique car model.

â€¢ ['Selling_Price'].mean() computes the average selling price for each group.

â€¢ .sort_values(ascending = False) sorts the results from highest to lowest average price.

-> Why this matters:

â€¢ Highlights which car models tend to sell for more

â€¢ Useful for identifying premium vs budget models

â€¢ Can guide pricing strategies, resale value analysis, or brand comparisons

```python
#Q11 Percentage of cars by fuel type
fuel_present = df['Fuel_Type'].value_counts(normalize = True) * 100
print("\n 25. Percentage of cars by Fuel_Type:\n", fuel_present)
```

-> Explanation:

â€¢ value_counts(normalize=True) returns the relative frequency of each fuel type as a proportion.

â€¢ Multiplying by 100 converts it into a percentage.

â€¢ The result shows what share of the dataset each fuel type occupies (e.g., Petrol: 60%, Diesel: 35%, CNG: 5%).

-> Why this matters:

â€¢ Reveals dominant fuel types in the dataset

â€¢ Useful for market trend analysis and segmentation

â€¢ Helps in comparing performance, pricing, or mileage across fuel types
